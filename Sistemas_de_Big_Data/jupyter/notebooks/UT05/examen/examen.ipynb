```python
import pandas as pd
import altair as alt

df = pd.read_csv("earthquake_1995-2023.csv")

display(df)
```

## Ejercicio 1

```python
grafico1 = ( alt
     .Chart(df)
     .mark_point()
     .encode(
         alt.X("magnitude:Q",scale=alt.Scale(zero=False)),
         alt.Y('depth:Q'),
         color='alert:N',
         size='sig:Q',
         tooltip = [
            alt.Tooltip('title:N'),
            alt.Tooltip('magnitude:Q'),
            alt.Tooltip('location:N')
        ] 
     )
    .properties(width=500, height=500)
)

grafico1
```

## Ejercicio 2

```python
linea = (alt
    .Chart(df)
    .mark_line()
    .encode(
        alt.Y('max(magnitude):Q',scale=alt.Scale(zero=False)),
        x='year(date_time):T',
    ).properties(width=1200, height=500)
)

puntos = (alt
    .Chart(df)
    .mark_point()
    .encode(
        alt.Y('max(magnitude):Q',scale=alt.Scale(zero=False)),
        x='year(date_time):T',
    ).properties(width=1200, height=500)
)

grafico2 = linea + puntos
grafico2
```

## Ejercicio 3

```python
grafico3 = (alt.Chart(df)
            .mark_rect()
            .encode(
                alt.X('magnitude:Q' , bin=alt.BinParams()),
                alt.Y('depth:Q' , bin=alt.BinParams()),
                alt.Color('count()')                
            ).properties(width=550, height=500)
            .transform_filter(
                  'datum.sig > 500'
            )
)

grafico3
```

## Ejercicio 4

```python
grafico4 = (alt
    .Chart(df)
    .transform_filter(
        'datum.continent !="null"'
    )
    .mark_bar()
    .encode(
        x='magnitude:Q',
        y='count():Q',
        column='continent:N'
    )
    .properties(width=150, height=150)  

)

grafico4
```

## Ejercicio 5

```python

(grafico2 & (grafico1 | grafico3) & grafico4)
```
