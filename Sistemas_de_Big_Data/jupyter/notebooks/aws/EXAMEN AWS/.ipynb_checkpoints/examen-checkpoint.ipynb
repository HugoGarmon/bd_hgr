## Programo la funcion para que envia el contenido a la cola

```python
import json
import urllib.parse
import boto3

s3 = boto3.client("s3")
sqs = boto3.client("sqs")


def lambda_handler(event, context):

    bucket = event["Records"][0]["s3"]["bucket"]["name"]
    nombre_archivo = urllib.parse.unquote_plus(event['Records'][0]['s3']['object']['key'])

    respuesta = s3.get_object(Bucket=bucket, Key=nombre_archivo)
    
    contenido = respuesta["Body"].read().decode("utf-8")
    contenido = contenido.split("\n")
    
    for i in range(len(contenido)):
        if i == 0:
            print("Saltando cabecera")
            contenido[i]
        else:
            respuesta = sqs.send_message(
            QueueUrl="https://sqs.us-east-1.amazonaws.com/381491882941/colaexamen",
            MessageBody=contenido[i]
        )
    return {
        'statusCode': 200,
        'body' : contenido
    }

```

## Creo el codigo de la funcion lambda
```python
import json
import boto3

s3 = boto3.client('s3')

def lambda_handler(event, context):
    
    mensajeCompleto = event["Records"][0]["Sqs"]['Message']

    mensaje = mensajeCompleto.split(",")
    for i in range(len(mensaje)):
        if mensaje[i] == "ERR":
            print("Error de formato")
            return {
                'statusCode': 400,
                'body':  mensajeCompleto
                
            }

    s3.put_object(
        Bucket='smartagro-clean-data-hgr',
        Key ="datosClean.json",
        Body = json.dumps(mensajeCompleto)
    )
            

    return {
        'statusCode': 200,
        'body': mensaje
    }
```
