```python
import requests
import pandas as pd

url = "https://swapi.dev/api/vehicles/"

response = requests.get(url)

datos = response.json()

datos = datos["results"]

df = pd.json_normalize(datos)
df.head(5)
```

```python
url = "https://swapi.dev/api/people/"


lista_dfs = [] 

while url:
    response = requests.get(url)
    data = response.json()

    df_pagina = pd.json_normalize(data["results"])
    lista_dfs.append(df_pagina)
    

    url = data["next"]

df_final = pd.concat(lista_dfs, ignore_index=True)
df_final
```

```python
def info_planeta(url):
    response = requests.get(url)
    datos = response.json()
    return datos['name'], datos['terrain'], datos['population']


df_20 = df_final.head(20).copy()


nombres = []
terrenos = []
poblaciones = []


for url in df_20["homeworld"]:
    nombre, terreno, poblacion = info_planeta(url)
    nombres.append(nombre)
    terrenos.append(terreno)
    poblaciones.append(poblacion)



df_final.loc[:19, 'planet_name'] = nombres
df_final.loc[:19, 'planet_terrain'] = terrenos
df_final.loc[:19, 'planet_population'] = poblaciones

df_final.head(20)
```

```python
cache_peliculas = {}
 
def obtener_titulo_pelicula(url_film):

    if url_film not in cache_peliculas:
        response = requests.get(url_film)
        if response.status_code == 200:
            cache_peliculas[url_film] = response.json().get('title')
        else:
            return "Desconocida"
    return cache_peliculas[url_film]
 
df_films = df_final.copy()

df_films = df_films.explode('films')
 
df_films['films'] = df_films['films'].apply(obtener_titulo_pelicula)

print(df_films[['name', 'films']].head(10))
 
media_personajes = df_films.groupby('films')['name'].count().mean()

print(f"Media de personajes por película {media_personajes}")
 
```
