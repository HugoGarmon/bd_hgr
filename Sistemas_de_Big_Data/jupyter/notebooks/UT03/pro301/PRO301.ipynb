```python
import pandas as pd
import json
```

```python
df_norte = pd.read_csv("./ventas_norte.csv",sep=";")
df_sur = pd.read_excel("./ventas_sur.xlsx", sheet_name=None)
df_sur = pd.concat(df_sur,ignore_index=True)

with open('ventas_este.json', 'r', encoding='utf-8') as f:
    datos_crudos = json.load(f)

df_este = pd.json_normalize(datos_crudos)
```

```python
nombres_estandar = {
    'Fecha_Venta': 'fecha',
    'Nom_Producto': 'producto',
    'Cantidad_Vendida': 'cantidad',
    'Precio_Unit': 'precio_unitario'
}

df_norte = df_norte.rename(columns=nombres_estandar)
df_norte['region'] = 'Norte'
df_norte
```

```python
mapping_sur = {
    'Date': 'fecha',
    'Item': 'producto',
    'Qty': 'cantidad',
    'UnitPrice': 'precio_unitario'
}

df_sur = df_sur.rename(columns=mapping_sur)
df_sur['region'] = 'Sur'

columnas_finales = ['fecha', 'producto', 'cantidad', 'precio_unitario', 'region']
df_sur = df_sur[columnas_finales]

df_sur
```

```python
df_este = df_este.rename(columns={
    'timestamp': 'fecha',
    'detalles_producto.nombre': 'producto',
    'detalles_producto.specs.cantidad': 'cantidad',
    'detalles_producto.specs.precio': 'precio_unitario'
})

df_este['region'] = 'Este'
df_este = df_este[['fecha', 'producto', 'cantidad', 'precio_unitario', 'region']]

df_este
```

```python
df_total = pd.concat([df_norte,df_este,df_sur],ignore_index=True)
```

```python
df_total.to_csv("ventas_consolidadas_2023.csv", sep=",", encoding="utf-8", index=False)
```
